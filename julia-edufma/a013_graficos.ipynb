```julia
# Graficos em Julia ...
using Pkg
Pkg.add("Plots")
```

```julia
# Grafico de linhas
using Plots
x = 1:10
y = rand(10)
z = rand(10)
plot(x,y)
plot!(x,z)
```

```julia
# Salvar grafico
x = 0:0.01:2*pi
y = sin.(x)
plot(x,y)
#savefig("seno.png")
```

```julia
# Exemplo mais completo ...
using Plots
x = 0:0.01:2*pi
y = sin.(x)
z = cos.(x)
plot(x,y)
plot!(x,z)
plot!(title="Funcoes Senoidais")
plot!(ylabel="Eixo Y")
plot!(xlabel="Eixo X")
```

```julia
using Plots

# Gerando alguns dados de exemplo
x = 1:10
y1 = x .^ 2  # y1 = x^2
y2 = 2 .* x .+ 3  # y2 = 2x + 3

# Criando a figura com dois gráficos e configurando as legendas, marcadores, etc.
plot(x, y1, label="y = x^2", marker=:circle, linewidth=2, color=:blue, xlabel="Eixo X", ylabel="Eixo Y", title="Gráficos de y = x^2 e y = 2x + 3")
plot!(x, y2, label="y = 2x + 3", marker=:square, linewidth=2, color=:red)

# Exibindo o gráfico
#display(plot)
```
