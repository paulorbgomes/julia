```julia
using Plots
```

```julia
# Data generation
year = [x for x in range(2000, length=10)];
sales = 100 * rand(10);
```

```julia
# Plot function
plot(year, sales, labels="Line Plot", color=:blue)
#plot(year, sales, labels="Line Plot", color=:blue, marker=:o, linewidth=2, linestyle=:dash)
scatter!(year, sales, labels="Scatter Plot", color=:blue)
title!("Sales x Year")
xlabel!("Year")
ylabel!("Sales")
```

```julia
# Data genaration (from a math function)
x = range(-10,length=21)
y = x.^2
plot(x, y, labels="Squared values", color=:red, linestyle=:dash, linewidth=2, marker=:o)
xlabel!("X")
ylabel!("Y")
title!("Y = X^2")
```

```julia
# Multiple graphs
x = range(-5, step=0.01, 5)
y = x.^2

p1 = plot(x, x, linewidth=2)
p2 = plot(y, x, linewidth=2)
p3 = plot(sin, x, linewidth=2)
p4 = plot(cos, x, linewidth=2)
plot(p1, p2, p3, p4, layout=(1,4))
```
